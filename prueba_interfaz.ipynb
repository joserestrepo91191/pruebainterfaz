```python
!pip install ipywidgets

import ipywidgets as widgets
from IPython.display import display

titulo = widgets.HTML('<h2>Interfaz gráfica para compartir</h2>')

descripcion = widgets.HTML('<p>Introduce tu nombre:</p>')

nombre = widgets.Text(value='', description='Nombre:', disabled=False)

boton_saludar = widgets.Button(description='Saludar')
boton_saludar.style.button_color = 'lightblue'

def saludar(b):
    nombre_persona = nombre.value
    
    with output:
        output.clear_output()
        print(f'Hola {nombre_persona}!')

output = widgets.Output()

boton_saludar.on_click(saludar)

input_container = widgets.VBox([descripcion, nombre, boton_saludar])

output_container = widgets.VBox([titulo, output])

display(widgets.HBox([input_container, output_container]))
```
